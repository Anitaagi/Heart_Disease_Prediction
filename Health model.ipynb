```python
# Replace 14+ with 14.0 and ensured all values are floats
# errors='coerce' will turn any other strings to NaN temporarily
```

```python
df['Holding_Policy_Duration'] = df['Holding_Policy_Duration'].replace('14+', 14.0)
df['Holding_Policy_Duration'] = pd.to_numeric(df['Holding_Policy_Duration'], errors='coerce')
```

```python
df.head(20)
```

```python
print(df['Holding_Policy_Duration'].isna().sum())
```

```python
print(df['Holding_Policy_Duration'].dtype)
```

```python
# check to see if NaN values mostly occur when policy type is also missing
print(df[df['Holding_Policy_Duration'].isna()]
    ['Holding_Policy_Type'].value_counts(dropna=False))
```

```python
# Drop the ID column (it's unique to each row and won't help the model learn)
df.drop('ID', axis=1, inplace=True)

# fill Duration with 0 and Type with a placeholder (0.0)
df['Holding_Policy_Duration'] = df['Holding_Policy_Duration'].fillna(0)
df['Holding_Policy_Type'] = df['Holding_Policy_Type'].fillna(0.0)

# Handle 'Health Indicator' missing values
# Since it's categorical (X1, X2, etc.), we fill with the most frequent value (Mode)
health_mode = df['Health Indicator'].mode()[0]
df['Health Indicator'] = df['Health Indicator'].fillna(health_mode)

# Convert Binary Strings to 0 and 1
# This makes the data readable for mathematical models
binary_mapping = {'Yes': 1, 'No': 0, 'Owned': 1, 'Rented': 0, 'Individual': 1, 'Joint': 0}

df['Is_Spouse'] = df['Is_Spouse'].map(binary_mapping)
df['Accomodation_Type'] = df['Accomodation_Type'].map(binary_mapping)
df['Reco_Insurance_Type'] = df['Reco_Insurance_Type'].map(binary_mapping)

# results
print(df.isnull().sum())
print(df.head(20))
```

```python
# One-Hot Encoding for remaining categorical columns
# We use drop_first=True to avoid the 'dummy variable trap' (multi-collinearity)
df_final = pd.get_dummies(df, columns=['City_Code', 'Health Indicator'], drop_first=True)

# Final check of the data types and shape
print(f"Original shape: {df.shape}")
print(f"New shape after encoding: {df_final.shape}")
print(df_final.head(20))
```
